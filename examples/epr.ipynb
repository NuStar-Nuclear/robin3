# Modeling 2D 1/8 EPR

```python
%matplotlib inline
from pprint import pprint
import numpy as np
import robin3
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
robin3.config.USE_TQDM_NOTEBOOK = True
```

## Loading Result from ROBIN3 Output HDF5 File

```python
result = robin3.ResultProcessor('EPR.h5')
```

display the summary of the simulation resultï¼š

```python
result.summary()
```

## Model Visualization

```python
plotter = result.generate_plotter()
```

```python
fig = plotter.plot_geometry()
```

## Result Visualization

### Print assembly power distribution:

```python
i = 0
for p in result.assembly_power():
    if p > 0:
        i += 1
        print(i, p)
```

### Plotting assembly power distribution:

```python
fig = plotter.plot_assembly_power()
```

### Plotting pin power distribution:

```python
fig = plotter.plot_pin_power()
```

### Plotting cell flux distribution:

```python
for g in [1, 2]:
    fig = plotter.plot_pin_flux(group=g)
```

### Plotting mesh flux distribution:

```python
for g in [1, 2]:
    fig = plotter.plot_mesh_flux(group=g)
```
